```python
from sklearn.linear_model import LinearRegression

# Convertir la columna 'datetime' a tipo datetime si no lo es
pot['datetime'] = pd.to_datetime(pot['datetime'])

# Años a proyectar
projection_years = [2035, 2040]

# Crear un DataFrame para la proyección
projection_df = pd.DataFrame({'datetime': pd.to_datetime(projection_years, format='%Y')})

# Realizar la proyección utilizando un modelo de regresión lineal
for column in ['eolica', 'solar']:
    X = pot['datetime'].dt.year.values.reshape(-1, 1)
    y = pot[column].values.reshape(-1, 1)
    
    # Crear el modelo de regresión lineal y ajustarlo a los datos existentes
    model = LinearRegression()
    model.fit(X, y)
    
    # Realizar la proyección para los años deseados
    X_projection = np.array(projection_years).reshape(-1, 1)
    projected_values = model.predict(X_projection)
    
    # Agregar las columnas proyectadas al DataFrame de proyección
    projection_df[column] = projected_values.flatten()

# Concatenar los DataFrames original y proyectado
pot_proyeccion = pd.concat([pot, projection_df], ignore_index=True)

# Mostrar el DataFrame con la proyección
print(pot_proyeccion)
```

```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d

# Datos existentes
existing_years = [2015, 2020, 2025, 2030]
existing_eolica_values = pot['eolica']
existing_solar_values = pot['solar']

# Crear funciones de interpolación lineal para cada variable dependiente
interp_eolica = interp1d(existing_years, existing_eolica_values, kind='linear')
interp_solar = interp1d(existing_years, existing_solar_values, kind='linear')

# Generar los valores proyectados para los años 2035 y 2040
projected_years = [2035, 2040]
projected_eolica_values = interp_eolica(projected_years)
projected_solar_values = interp_solar(projected_years)

# Crear un nuevo dataframe con los valores proyectados
projected_pot = pd.DataFrame({'datetime': projected_years,
                              'eolica': projected_eolica_values,
                              'solar': projected_solar_values})

projected_pot.head(6)
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit

# Datos existentes
existing_years = [2015, 2020, 2025, 2030]
existing_eolica_values = pot['eolica']
existing_solar_values = pot['solar']

# Definir una función polinómica de grado 2
def polynomial_func(x, a, b, c):
    return a * x**2 + b * x + c

# Ajustar la curva a los datos existentes para cada variable dependiente
coefficients_eolica, _ = curve_fit(polynomial_func, existing_years, existing_eolica_values)
coefficients_solar, _ = curve_fit(polynomial_func, existing_years, existing_solar_values)

# Generar los valores proyectados para los años 2035 y 2040
projected_years = [2035, 2040]
projected_eolica_values = polynomial_func(projected_years, *coefficients_eolica)
projected_solar_values = polynomial_func(projected_years, *coefficients_solar)

# Crear un nuevo dataframe con los valores proyectados
projected_pot = pd.DataFrame({'datetime': projected_years,
                              'eolica': projected_eolica_values,
                              'solar': projected_solar_values})

projected_pot.head(6)
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit

# Datos existentes
existing_years = [2015, 2020, 2025, 2030]
existing_eolica_values = pot['eolica']
existing_solar_values = pot['solar']

# Definir una función polinómica de grado 2
def polynomial_func(x, a, b, c):
    return a * x**2 + b * x + c

# Ajustar la curva a los datos existentes para cada variable dependiente
coefficients_eolica, _ = curve_fit(polynomial_func, existing_years, existing_eolica_values)
coefficients_solar, _ = curve_fit(polynomial_func, existing_years, existing_solar_values)

# Generar los valores proyectados para los años 2035 y 2040
projected_years = [2035, 2040]
projected_eolica_values = [polynomial_func(year, *coefficients_eolica) for year in projected_years]
projected_solar_values = [polynomial_func(year, *coefficients_solar) for year in projected_years]

# Crear un nuevo dataframe con los valores proyectados
projected_pot = pd.DataFrame({'datetime': projected_years,
                              'eolica': projected_eolica_values,
                              'solar': projected_solar_values})

projected_pot.head(6)
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit

# Datos existentes
existing_years = [2015, 2020, 2025, 2030]

# Obtener las columnas de variables dependientes del dataframe
columns = ['eolica', 'solar_FV', 'solar_termoelectrica', 'hidraulica', 'bombeo_mixto', 'bombeo_puro', 'biogas',
           'otras_renovables', 'biomasa', 'carbon', 'ciclo_combinado', 'cogeneracion', 'fuel_gas',
           'residuos_otros', 'nucelar', 'almacenamiento']

# Ajustar la curva y generar las proyecciones para cada columna de variables dependientes
projections = {}
for column in columns:
    existing_values = pot[column]

    # Ajustar la curva a los datos existentes para la columna actual
    coefficients, _ = curve_fit(polynomial_func, existing_years, existing_values)

    # Generar los valores proyectados para los años 2035 y 2040 para la columna actual
    projected_values = [polynomial_func(year, *coefficients) for year in projected_years]

    # Almacenar los valores proyectados en el diccionario de proyecciones
    projections[column] = projected_values

# Crear un nuevo dataframe con los valores proyectados
projected_pot = pd.DataFrame({'datetime': projected_years, **projections})

projected_pot.head(6)
```
