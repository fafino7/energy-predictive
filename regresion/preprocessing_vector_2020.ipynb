```python
import pandas as pd

vector = pd.read_csv('/content/drive/MyDrive/Data/energy/regresion/raw_vector_2020.csv', delimiter=';')
vector.head()
```

```python
# Verificar y convertir las columnas de potencia energética a valores numéricos
power_columns = ['NR', 'solar', 'eolica', 'otras_ren', 'otras']
for column in power_columns:
    vector[column] = pd.to_numeric(vector[column], errors='coerce')
```

El df no tiene el atributo str directamente, por lo que no se puede utilizar el método str.replace() en todo el df de una vez.

Así que se aplicó el método 'str.replace()' únicamente a las columnas de % individualmente utilizando un bucle 'for' para recorrer las columnas.

```python
percentage_columns = ['%NR', '%solar', '%eolica', '%otras_ren', '%otras']

for column in percentage_columns:
    vector[column] = vector[column].astype(str).str.replace('%', '')  # Eliminar el símbolo de porcentaje
    vector[column] = vector[column].astype(str).str.strip()  # Eliminar espacios en blanco adicionales
```

```python
for column in percentage_columns:
    vector[column] = vector[column].astype(str).str.replace(',', '.')  # Eliminar el símbolo de porcentaje
    vector[column] = vector[column].astype(str).str.strip()  # Eliminar espacios en blanco adicionales
```

```python
# Convertir las columnas de distribución porcentual a valores numéricos
for column in percentage_columns:
    vector[column] = pd.to_numeric(vector[column], errors='coerce')
```

```python
# Verificar y ajustar los valores de las columnas de distribución porcentual para que sumen 1
total_percentage = vector['%total'].sum()
if total_percentage != 1:
    vector[percentage_columns] = vector[percentage_columns] / total_percentage
```
